```python
instance=[['sunny','warm','normal','strong','warm','same'],
          ['sunny','warm','high','strong','warm','same'],
          ['rainy','cold','high','strong','warm','change'],
          ['sunny','warm','high','strong','cool','change']]
target=['yes','yes','no','yes']
print("specific hyp")
sh=['null']*len(instance[0])
print("general hyp")
gen = [["?" for i in range(len(sh))] for i in range(len(sh))]
'''gen=[['?','?','?','?','?','?'],['?','?','?','?','?','?']
     ,['?','?','?','?','?','?'],['?','?','?','?','?','?']
     ,['?','?','?','?','?','?']['?','?','?','?','?','?']]'''
i=0
while(i<len(instance)):
    if (target[i]=='yes'):
        if sh[0]=='null':
            sh=instance[0].copy()
        else:
            for x in range(len(sh)):
                if sh[x]!=instance[i][x]:
                    sh[x]='?'
    #no condition
    else:
        for x in range(len(instance[0])):
            if sh[x]!=instance[i][x]:
                gen[x][x]=sh[x]
            else:
                gen[x][x]='?'
    print("specific hyp\n")
    print(sh)
    print("genral hyp\n")
    for g in gen:
            print(g)
                
    i+=1
    
for x in range(len(instance[0])):
    if (gen[x][x]!=sh[x]):
        gen[x][x]='?'

print("general hyp is \n")
for a in gen:
    flag=0
    for s in a:
        if s!='?':
            flag=1
            break
    if flag==1:
        print(a)

            
        
        
```
